#Exercise 1

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft, ifft

N = 64
tau = 10.0
t = np.linspace(0, tau, N, endpoint=False)
print(t)
f = 0.2
x = np.cos(2 * np.pi * f * t)

plt.figure()
plt.plot(t, x)
plt.plot(t, x, 'o')
plt.xlabel('Time')
plt.ylabel('x(N)')
plt.title('Single cosine wave sampled at 64 points')
plt.grid()
plt.show()
```

#Exercise 2

```python
X = fft(x)/N

plt.figure()
plt.plot(np.abs(X)[0:32])
plt.plot(np.abs(X)[0:32], 'o')
plt.xlabel('Point number')
plt.ylabel('X(N)')
plt.title('Positive frequencies of FFT of above cosine wave sampled at 64 points')
plt.grid()
plt.show()
```

# Exercise 3

```python
def spectrum1(x, Fs):
  N = len(x)
  freq = (float(Fs)/N)*np.arange(0, N/2 + 1)
  X = fft(x)/N
  X = X[0:N//2 + 1]
  return freq, X

Fs = N / tau

freq_result, X_result = spectrum1(x, Fs)

plt.figure()
plt.plot(freq_result, np.abs(X_result))
plt.plot(freq_result, np.abs(X_result), 'o')
plt.xlabel('Frequency')
plt.ylabel('Amplitude of X(N)')
plt.title('FFT of x vs frequency')
plt.grid()
plt.show()
```

The signal amplitude was equal to 0.5 instead of 1 because it's been normalised

#Exercise 4

```python
N = 64
tau = 10.0
t = np.linspace(0, tau, N, endpoint=False)
print(t)
f1 = 0.2
x1 = np.cos(2 * np.pi * f1 * t)
f2 = 0.22
x2 = np.cos(2 * np.pi * f2 * t)

plt.figure()
plt.plot(t, x1)
plt.plot(t, x1, 'o', label='Frequency = 0.2')
plt.plot(t, x2)
plt.plot(t, x2, 'o', label='Frequency = 0.22')
plt.xlabel('Time')
plt.ylabel('x(N)')
plt.title('Single cosine wave sampled at 64 points')
plt.legend()
plt.grid()
plt.show()
```

I see that the wavelength decreases as the frequency increases

```python
N = 64
tau1 = 10.0
tau2 = 25
t1 = np.linspace(0, tau1, N, endpoint=False)
t2 = np.linspace(0, tau2, N, endpoint=False)
f1 = 0.2
f2 = 0.22
x1 = np.cos(2 * np.pi * f1 * t1)
x2 = np.cos(2 * np.pi * f1 * t2)
x3 = np.cos(2 * np.pi * f2 * t1)
x4 = np.cos(2 * np.pi * f2 * t2)

plt.figure()
plt.plot(t1, x1, label='tau = 10, f = 0.2')
plt.plot(t2, x2, 'g--', label='tau = 25, f = 0.2')
plt.plot(t1, x3, label='tau = 10, f = 0.22')
plt.plot(t2, x4, 'r--', label='tau = 25, f = 0.22')
plt.xlabel('Time')
plt.ylabel('x(N)')
plt.title('Single cosine wave sampled at 64 points')
plt.legend()
plt.grid()
plt.show()
```

As tau increases, the time period increases by the same amount
